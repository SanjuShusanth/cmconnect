```python
import matplotlib.pyplot as plt
```

```python
df_grouped = df.groupby('district')['grievance_id'].count().reset_index()

plt.bar(df_grouped['district'], df_grouped['grievance_id'])
plt.xlabel('District')
plt.ylabel('Number of Grievances')
plt.title('Grievances by District')
plt.xticks(rotation=45)
plt.show()

```
